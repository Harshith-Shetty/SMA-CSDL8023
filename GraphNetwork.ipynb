```python
#Social Networks Chapter 2 of Networkx #Modelling Road Network
```

```python
G=nx.Graph()       #Undirected Graph  G=nx.DiGraph() For Directed Graph
```

```python
city_set=['Delhi', 'Banglore', 'Hydrabad', 'Ahmadabad', 'Chennai', 'Kolkata', 'Surat','Pune', 'Jaipur']
```

```python
for each in city_set: 
  G.add_node(each)
nx.draw(G,with_labels=1)
```

```python
costs=[]
value=100
while (value<=2000):
  costs.append(value)
  value=value+100
```

```python
costs
```

```python
import random
#Adding 16 Edges to the network with weights
while(G.number_of_edges()<16):
    c1=random.choice(list(G.nodes()))
    c2=random.choice(list(G.nodes()))
    if (c1!=c2)and (G.has_edge(c1,c2))==0:
        w=random.choice(costs)
        G.add_edge(c1,c2,weight=w)
pos=nx.circular_layout(G)
#pos=nx.spectral_layout(G)
#pos=nx.spring_layout(G)
nx.draw(G,pos,with_labels=1)
nx.draw_networkx_edge_labels(G,pos)
```

```python
nx.is_connected(G)
```

```python
for u in G.nodes():
  for v in G.nodes():
    print(u,v,nx.has_path(G,u,v))
```
